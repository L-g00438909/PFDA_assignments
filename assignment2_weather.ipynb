## Dry Bulb Temperature(&deg;C) vs Time



```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter         # DateFormatter to format the date on the x-axis. 

# Load the dataset from a CSV file.
df = pd.read_csv('Weatherreadings1.csv')
```

**Data Manipulation:**

```python
df.head()
```

```python
df.info()
```

```python
# column names and number of columns
df.columns ,len(df.columns)
```

```python
# isolate reportStartDatetime and wetbulbtemperature_Celsius columns.
df = df[['reportStartDateTime', 'dryBulbTemperature_Celsius']]
df


```

**Data Analysis:**

```python


# Convert the 'reportStartDateTime' column to datetime format 
df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime'])

# setup the plot 
fig, ax = plt.subplots(figsize=(12, 8)) 

x = df["reportStartDateTime"]   # x-axis values
y = df["dryBulbTemperature_Celsius"]  # y-axis values 

# Create the plot
ax.plot(x, y, marker='o', color='blue', label="Temperature (Â°C)")  # marker is the shape of the data points. 

# Set the date format for the x-axis labels
ax.xaxis.set_major_formatter(DateFormatter("%b %d %H:%M"))         # %b is the month, %d is the day, %H is the hour, %M is the minute.

# Rotate and align the x-axis labels
plt.xticks(rotation=45, ha='right') #xticks are the x-axis labels. ha is the horizontal alignment. 

# Add a title and labels
ax.set_title("Dry Bulb Temperature (Â°C) Over Time", fontsize=20, fontweight='bold', color='black')
ax.set_xlabel("Time", fontsize=12, fontweight='bold', color='black')
ax.set_ylabel("Temperature (Â°C)", fontsize=12, fontweight='bold', color='black')

# Add a grid for better readability
ax.grid(True, linestyle='--', color='black', linewidth=0.5)

# Add a legend
ax.legend()

# Ensure everything fits well
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# max temperature 
max_temp = df['dryBulbTemperature_Celsius'].max()
print(max_temp)

# min temperature
min_temp = df['dryBulbTemperature_Celsius'].min()
print(min_temp)
```
