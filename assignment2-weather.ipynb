```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
# format date on x-axis
from matplotlib.dates import DateFormatter       

# Load the dataset from a CSV file.
df = pd.read_csv('Weatherreadings1.csv')
```

```python
# column names and number of columns
df.columns ,len(df.columns)
```

```python
# isolate reportStartDatetime and drybulbtemperature_Celsius columns.
df = df[['reportStartDateTime', 'dryBulbTemperature_Celsius']]
df


```

```python


# Convert the 'reportStartDateTime' column to datetime format 
df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime'])

# setup the plot 
fig, ax = plt.subplots(figsize=(12, 8)) 

x = df["reportStartDateTime"]   # x-axis values
y = df["dryBulbTemperature_Celsius"]  # y-axis values 

# Create the plot
ax.plot(x, y, marker='o', color='b', label="Temperature (°C)")  # marker is the shape of the data points. 

# Set the date format for the x-axis labels
ax.xaxis.set_major_formatter(DateFormatter("%b %d %H:%M"))         # %b is the month, %d is the day, %H is the hour, %M is the minute.

# Rotate and align the x-axis labels
plt.xticks(rotation=45, ha='right') #xticks are the x-axis labels. ha is the horizontal alignment. 

# Add a title and labels
ax.set_title("Dry Bulb Temperature (°C) vs Time", fontsize=20, fontweight='bold', color='black')
ax.set_xlabel("Time", fontsize=12, fontweight='bold', color='black')
ax.set_ylabel("Temperature (°C)", fontsize=12, fontweight='bold', color='black')


max_temp = y.max()
min_temp = y.min()
max_time = x[y.idxmax()] # time of the maximum temperature
min_time = x[y.idxmin()] # time of the minimum temperature

# Format the max and min temperatures to 2 decimal places
max_temp_str = f"{max_temp:.2f}" 
min_temp_str = f"{min_temp:.2f}"

# Annotate the maximum temperature with 2 decimal places
ax.annotate(f"Max: {max_temp_str}°C",       
            xy=(max_time, max_temp), 
            xytext=(max_time, max_temp + 1),  # Position the annotation above the max point. +1 is the distance from the point.
            arrowprops=dict(facecolor='red'),
            fontsize=12, color='red',fontweight='bold', ha='center')

# Annotate the minimum temperature with 2 decimal places
ax.annotate(f"Min: {min_temp_str}°C", 
            xy=(min_time, min_temp), 
            xytext=(min_time, min_temp - 2),  # Position the annotation below the min point. -2 is the distance from the point.
            arrowprops=dict(facecolor='red'),
            fontsize=12, color='red', fontweight='bold', ha='center')


ax.grid(True, linestyle='--', color='black', linewidth=0.5)
ax.legend()
plt.tight_layout()
fig.savefig('Temperature_Plot.png')  
plt.show()

```

```python
# max temperature
max_temp = df['dryBulbTemperature_Celsius'].max()
print(f'\033[1mMax Temp:\033[0m {max_temp}')

# max temperature time
max_temp_time = df.loc[df['dryBulbTemperature_Celsius'].idxmax()]['reportStartDateTime']
print(f'\033[1mMax Temp time:\033[0m {max_temp_time}')

# min temperature 
min_temp = df['dryBulbTemperature_Celsius'].min()
print(f'\033[1mMin Temp:\033[0m {min_temp}')

# min temperature time
min_temp_time = df.loc[df['dryBulbTemperature_Celsius'].idxmin()]['reportStartDateTime']
print(f'\033[1mMin Temp time:\033[0m {min_temp_time}')
```

     
**The minimum dry bulb temperature was 9.17°C recorded on 03/10/24 at 00:15, while the maximum temperature was 15.94°C recorded on 02/10/24 at 16:15.**
